```python
#  Defining main category
df['MainCategory'] = df['product_category_tree'].apply(lambda x: x.split('>>')[0][2:])
```

```python
# Dividing data into categories
def secondary(x):
    try:
        return x.split('>>')[1][1:]
    except IndexError:
        return 'None '
    
def tertiary(x):
    try:
        return x.split('>>')[2][1:]
    except IndexError:
        return 'None '
    
def quaternary(x):
    try:
        return x.split('>>')[3][1:]
    except IndexError:
     return 'None '
```

```python
# Dividing data into various category based on their type. 
df['SecondaryCategory'] = df['product_category_tree'].apply(secondary)
df['TertiaryCategory'] = df['product_category_tree'].apply(tertiary)
df['QuaternaryCategory'] = df['product_category_tree'].apply(quaternary)
```
